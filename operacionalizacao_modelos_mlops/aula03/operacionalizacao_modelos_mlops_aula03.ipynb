# Operacionalização de Modelos com MLOps

## Aula 03: Fundamentos de Qualidade de Dados

1. Introdução: O Fim da Ingestão e o Começo da Inteligência

Vocês passaram as últimas aulas aprendendo a extrair dados de fontes diversas — como buckets S3, bancos SQL e arquivos Parquet. Mas o que acontece quando esses dados chegam? O paradigma agora muda de "como transportamos o dado" para "como transformamos esse dado em uma decisão".

Sem dúvida, a qualidade dos dados é o elemento mais importante para alcançar resultados bem-sucedidos em modelos de ML. Muitos cursos ignoram essa parte por ser considerada "tediosa", mas o ditado Garbage In, Garbage Out (Lixo entra, Lixo sai) é a lei absoluta aqui. Se o seu pipeline de ingestão traz dados ruidosos ou enviesados, a capacidade do seu modelo de generalizar para novos dados desaparece, levando a decisões de negócio errôneas.

### 2. Fundação de Dados: Tratamento Técnico e Decisão Estratégica
Uma vez ingeridos, os dados raramente estão prontos. Precisamos de uma Fundação de Dados sólida.

#### 2.1. O Desafio dos Dados Ausentes

Dados faltantes surgem por erros de coleta, falhas de sistema ou corrupção
. Ignorar esses registros pode enviesar o entendimento do padrão subjacente pelo modelo
. No scikit-learn, temos três estratégias principais:
    - SimpleImputer: Substitui valores por estatísticas simples como média ou mediana. É leve e ideal quando menos de 5% dos dados estão ausentes.
    - KNNImputer: Usa o algoritmo de vizinhos mais próximos para "adivinhar" o valor com base em amostras similares. Captura relações mais sutis.
    - IterativeImputer: Uma abordagem avançada que modela cada característica com valores ausentes como uma função de outras características. É computacionalmente mais complexo, mas oferece predições mais precisas para os valores faltantes


#### 2.2. Escalonamento (Scaling): Por que os números importam?

Se você tem uma característica "Idade" (0-100) e outra "Renda" (0-100.000), algoritmos baseados em distância, como KNN, serão dominados pela Renda.
- StandardScaler: Transforma os dados para terem média 0 e desvio padrão 1 (Z-score). Essencial para dados com distribuição Gaussiana.
- MinMaxScaler: Redimensiona para um intervalo fixo, geralmente. Preserva o relacionamento entre os valores originais.

### 3. Lidando com Variáveis Categóricas e o ColumnTransformer

A maioria dos algoritmos de ML exige entrada numérica. Aqui, diferenciamos variáveis Nominais (sem ordem, ex: cor) de Ordinais (com ordem, ex: cargo Junior/Senior).
Para lidar com datasets mistos, usamos o ColumnTransformer(). Ele permite aplicar técnicas diferentes em colunas diferentes simultaneamente. Por exemplo, você pode aplicar StandardScaler em colunas numéricas e OneHotEncoder em colunas categóricas no mesmo passo.

### 4. O Coração da Operacionalização: Scikit-learn Pipelines

Um pipeline no scikit-learn é uma sequência de etapas executadas em ordem. Por que isso é vital para a operacionalização?

- Consistência: Garante que as mesmas transformações aplicadas no treino sejam aplicadas na predição, evitando erros manuais.
- Prevenção de Data Leakage (Vazamento de Dados): O vazamento ocorre quando informações de fora do conjunto de treino são usadas para criar o modelo. Ao usar pipelines com train_test_split, garantimos que o escalonamento, por exemplo, aprenda a média apenas com os dados de treino

### 5. Experimentação Avançada: Feature Engineering e Seleção

Não basta usar o que veio da ingestão. A Feature Engineering permite extrair mais sinal dos dados.

- PolynomialFeatures: Cria termos de interação entre variáveis (ex: x1⋅x2), o que pode revelar padrões não lineares.
- Redução de Dimensionalidade e Seleção: Modelos mais simples são preferíveis (Navalha de Occam). 
- O RFE (Recursive Feature Elimination) remove recursivamente as características menos importantes com base no ranking de um estimador. Isso aumenta a interpretabilidade e reduz o overfitting

### 6. Conclusão e Impacto de Negócio

Um projeto de ML bem estruturado não termina no .fit(). Ele deve ser modular para permitir o rastreamento de experimentos (como o feito com MLflow, embora o foco aqui seja a estrutura do scikit-learn) e a fácil substituição de componentes.

Ao automatizar o pipeline desde a fundação de dados até a validação, o cientista de dados foca na tomada de decisão técnica: "Qual estratégia de imputação minimiza meu erro?" ou "Quais features realmente impactam o lucro da empresa?"
