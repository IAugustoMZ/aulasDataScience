## Introdução a Modelos Lineares para Regressão e Classificação

### Introdução aos Modelos Lineares

No campo do Machine Learning, os **modelos lineares** são a espinha dorsal de muitos algoritmos de regressão e classificação. Eles são chamados de 'lineares' porque a saída é uma função linear de suas entradas. Em termos mais simples, eles tentam encontrar uma relação de linha reta, plano ou hiperplano para descrever a conexão entre variáveis.

#### Importância Fundamental no Machine Learning

A importância fundamental dos modelos lineares reside em sua **simplicidade, interpretabilidade e eficiência computacional**. Eles servem como um excelente ponto de partida para a modelagem preditiva e são frequentemente usados como modelos de *baseline* para comparar algoritmos mais complexos. Além disso, muitos algoritmos mais avançados incorporam conceitos lineares em sua estrutura, tornando a compreensão dos modelos lineares essencial para o domínio do Machine Learning.

#### Aplicações em Machine Learning

Modelos lineares são incrivelmente versáteis e têm uma vasta gama de aplicações no mundo real:

*   **Previsão de Preços de Casas:** Estimar o valor de uma propriedade com base em características como tamanho, número de quartos e localização.
*   **Classificação de E-mails como Spam:** Decidir se um e-mail é spam ou não com base em palavras-chave e padrões no conteúdo.
*   **Análise de Mercado Financeiro:** Prever o preço de ações ou outros ativos com base em tendências históricas e indicadores econômicos.
*   **Diagnóstico Médico:** Prever a probabilidade de uma doença com base em sintomas e dados do paciente.

3.  **Regressão Linear:** O modelo clássico para prever um valor contínuo com base em uma ou mais variáveis de entrada.

### O que é um Modelo Linear?

Um **modelo linear** é um tipo de algoritmo de aprendizado de máquina que assume uma relação linear entre as variáveis de entrada (features) e a variável de saída (target). Em outras palavras, ele tenta encontrar uma linha, um plano ou um hiperplano (em dimensões superiores) que melhor se ajuste aos dados para fazer previsões ou classificações.

**Como funciona?**

Imagine que você tem dados de pontos em um gráfico 2D. Um modelo linear de regressão tentaria desenhar uma linha reta que passa o mais próximo possível da maioria desses pontos. Para novos dados, a previsão seria simplesmente o valor na linha reta correspondente à entrada. Para classificação, um modelo linear tentaria desenhar uma linha que separa os pontos de diferentes categorias.

**Vantagens Principais:**

1.  **Simplicidade:** Modelos lineares são relativamente fáceis de entender e implementar. Sua estrutura direta os torna um bom ponto de partida para muitos problemas de Machine Learning.
2.  **Interpretabilidade:** A relação linear significa que podemos entender facilmente como cada variável de entrada afeta a saída. Por exemplo, em uma regressão, o coeficiente de uma variável nos diz o quanto a saída muda para cada unidade de aumento nessa variável, mantendo o resto constante. Isso os torna valiosos em cenários onde a explicação do modelo é tão importante quanto a previsão.
3.  **Eficiência Computacional:** São geralmente rápidos para treinar e fazer previsões, mesmo com grandes conjuntos de dados.

**Intuição Geométrica:**

*   **Regressão Linear:** Pense em encontrar a "melhor linha" (ou plano/hiperplano) que atravessa seus pontos de dados, minimizando a distância entre os pontos e a linha. Esta linha representa a tendência geral dos dados.
*   **Classificação Linear (e.g., Regressão Logística, SVM linear):** A ideia é encontrar uma "melhor linha divisória" (ou plano/hiperplano) que separa as diferentes classes de dados. Todos os pontos de um lado da linha pertencem a uma classe, e todos os pontos do outro lado pertencem a outra.

## Perceptron para Classificação Binária (Risco de Crédito)

O Perceptron é um algoritmo fundamental no campo do aprendizado de máquina, especialmente para problemas de classificação binária. Ele foi um dos primeiros modelos a serem desenvolvidos para simular a tomada de decisão de um neurônio biológico. Essencialmente, o Perceptron aprende a dividir o espaço de entrada em duas regiões distintas por meio de uma linha (em 2D), um plano (em 3D) ou um hiperplano (em dimensões superiores). Essa divisão é conhecida como **limite de decisão**.

### Como Funciona?

Imagine que você é um banco que precisa decidir se aprova ou não um pedido de crédito. Você analisa diversas características do cliente, como:

*   **Renda Mensal:** Maior renda pode significar menor risco.
*   **Histórico de Dívidas:** Muitas dívidas anteriores podem significar maior risco.
*   **Idade:** Pessoas em certas faixas etárias podem ter perfis de risco diferentes.

O Perceptron age de forma semelhante. Ele recebe esses *inputs* (as características do cliente), multiplica cada um por um *peso* (que indica a importância dessa característica para a decisão), soma esses valores e adiciona um *viés* (bias). O resultado dessa soma é então passado por uma *função de ativação* (geralmente uma função degrau) que decide se o crédito deve ser aprovado (saída 1) ou negado (saída 0).

$$ \text{saída} = \begin{cases} 1 & \text{se } \sum_{i=1}^{n} w_i x_i + b \ge 0 \\ 0 & \text{se } \sum_{i=1}^{n} w_i x_i + b < 0 \end{cases} $$

O algoritmo aprende ajustando esses pesos ($w_i$) e o viés ($b$) iterativamente, com base nos erros que comete durante o treinamento. Se ele prevê 'aprovar' e o crédito deveria ter sido 'negado', ele ajusta os pesos para ser menos propenso a aprovar na próxima vez, e vice-versa.

### Limite de Decisão

O objetivo final do Perceptron é encontrar um limite de decisão que separe as classes de forma eficaz. No exemplo do crédito, este limite separaria os clientes "bons pagadores" dos "maus pagadores". No entanto, é importante notar que o Perceptron é eficaz apenas para dados que são *linearmente separáveis*, ou seja, onde é possível traçar uma linha reta (ou hiperplano) para separar perfeitamente as duas classes.

```python
import warnings
import numpy as np
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import classification_report, precision_score, r2_score
from sklearn.linear_model import Perceptron, LogisticRegression, LinearRegression
from sklearn.preprocessing import RobustScaler, PowerTransformer, KBinsDiscretizer, OneHotEncoder
```

```python
# ignorar todos os warnings
warnings.filterwarnings('ignore')

# definir o random state
RANDOM_STATE = 13
```

```python
# montar o drive no Colab
drive.mount('/content/drive')
```

```python
# caminho para o arquivo
PATH='/content/drive/MyDrive/Pos_Graduacao/FundamentosMachineLearning/data/cs-training.csv'
data = pd.read_csv(PATH, index_col=0)
data.head()
```

```python
# verificando dados nulos
display(data.isna().sum())
```

```python
# separar x e y
target = 'SeriousDlqin2yrs'
x = data.drop(columns=[target])
y = data[[target]]
```

```python
# separar entre treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=RANDOM_STATE)
```

```python
# pipeline 1 - perceptron + todas as features (simpleImputer + RobustScaler)
pipeline1 = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler()),
    ('perceptron', Perceptron())
])

# treinando o pipeline
pipeline1.fit(x_train, y_train)
```

```python
# acessar os passos
pipeline1['scaler'].center_
```

```python
# previsão dos dados com pipeline
yhat_train_pipe1 = pipeline1.predict(x_train)
yhat_test_pipe1 = pipeline1.predict(x_test)
```

```python
def evaluate(y_train, y_test, yhat_train, yhat_test):
  print('Treinamento')
  print(classification_report(y_train, yhat_train))
  print('*'*150)
  print('Teste')
  print(classification_report(y_test, yhat_test))
```

```python
# pipeline 2 - perceptron + todas as features (simpleImputer + PowerTransformer no MonthlyIncome + RobustScaler)
idx_monthly_income = [list(x_train.columns).index('MonthlyIncome')]

col_transfomer = ColumnTransformer(
    transformers=[
        ('power_transf', PowerTransformer(), idx_monthly_income)
    ],
    remainder='passthrough'
)

# construindo o pipeline
pipeline2 = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('power_transf', col_transfomer),
    ('scaler', RobustScaler()),
    ('perceptron', Perceptron())
])

# treinando o pipeline
pipeline2.fit(x_train, y_train)

# previsão dos dados com pipeline
yhat_train_pipe2 = pipeline2.predict(x_train)
yhat_test_pipe2 = pipeline2.predict(x_test)
```

```python
# avaliação do novo pipeline
evaluate(y_train, y_test, yhat_train_pipe2, yhat_test_pipe2)
```

## Regressão Logística para Classificação Binária (Risco de Crédito)

A Regressão Logística é um dos algoritmos de classificação mais amplamente utilizados no Machine Learning. Apesar do nome "Regressão", ela é fundamentalmente um modelo de **classificação**, projetado para prever a probabilidade de uma instância pertencer a uma determinada classe, especialmente em problemas de classificação binária.

### Extensão do Perceptron e a Função Sigmoide

Podemos pensar na Regressão Logística como uma extensão do Perceptron. Enquanto o Perceptron produz uma saída binária (0 ou 1) diretamente a partir de uma soma ponderada, a Regressão Logística leva essa soma ponderada e a passa por uma **função sigmoide** (também conhecida como função logística). A função sigmoide "espreme" qualquer valor real para um valor entre 0 e 1, que pode ser interpretado como uma probabilidade.

$$ P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X_1 + ... + \beta_nX_n)}} = \frac{1}{1 + e^{-z}} $$

Onde $z = \beta_0 + \beta_1X_1 + ... + \beta_nX_n$ é a combinação linear das características de entrada (features) e seus respectivos pesos (coeficientes). O resultado, $P(Y=1|X)$, é a probabilidade de a instância pertencer à classe positiva (neste caso, `SeriousDlqin2yrs=1`).

### Analogia com Risco de Crédito e Limiar de Probabilidade

Voltando ao exemplo do risco de crédito:

*   O Perceptron faria uma decisão direta: "Este cliente tem risco" ou "Este cliente não tem risco".
*   A Regressão Logística, por outro lado, diria: "Este cliente tem **85% de probabilidade** de ser um mau pagador" ou "Este cliente tem **12% de probabilidade** de ser um mau pagador".

Para converter essa probabilidade em uma decisão binária, utilizamos um **limiar de probabilidade** (threshold). Se a probabilidade calculada for maior ou igual a esse limiar (comumente 0.5), classificamos a instância como a classe positiva (risco). Caso contrário, classificamos como a classe negativa (sem risco).

Essa abordagem probabilística é extremamente útil, pois permite não apenas a classificação, mas também uma medida da confiança do modelo em sua previsão, o que é crucial em cenários como o financeiro, onde as consequências das decisões são significativas.

```python
# pipeline 3 - reglog + MonthlyIncome, Age (simpleImputer + PowerTransformer no MonthlyIncome + KBinsDiscretizer no Age + RobustScaler)
idx_monthly_income = [list(x_train.columns).index('MonthlyIncome')]
idx_age = [list(x_train.columns).index('age')]

col_transfomer = ColumnTransformer(
    transformers=[
        ('power_transf', PowerTransformer(), idx_monthly_income),
        ('kbins_disc', KBinsDiscretizer(n_bins=3), idx_age)
    ],
    remainder='drop'
)

# construindo o pipeline
pipeline3 = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('power_transf', col_transfomer),
    ('scaler', RobustScaler()),
    ('reglog', LogisticRegression())
])

# treinando o pipeline
pipeline3.fit(x_train, y_train)

# previsão dos dados com pipeline
yhat_train_pipe3 = pipeline3.predict(x_train)
yhat_test_pipe3 = pipeline3.predict(x_test)
```

```python
# avaliação do novo pipeline
evaluate(y_train, y_test, yhat_train_pipe3, yhat_test_pipe3)
```

```python
def evaluate_k(k, x_train, y_train, x_test, y_test):
  idx_monthly_income = [list(x_train.columns).index('MonthlyIncome')]
  idx_age = [list(x_train.columns).index('age')]

  col_transfomer = ColumnTransformer(
      transformers=[
          ('power_transf', PowerTransformer(), idx_monthly_income),
          ('kbins_disc', KBinsDiscretizer(n_bins=k), idx_age)
      ],
      remainder='drop'
  )

  # construindo o pipeline
  pipeline = Pipeline([
      ('imputer', SimpleImputer(strategy='median')),
      ('power_transf', col_transfomer),
      ('scaler', RobustScaler()),
      ('reglog', LogisticRegression())
  ])

  # treinando o pipeline
  print(f'Treinamento Pipeline com k = {k}')
  pipeline.fit(x_train, y_train)

  # previsão dos dados com pipeline
  yhat_train = pipeline.predict(x_train)
  yhat_test = pipeline.predict(x_test)

  # calculando precisão
  prec_train = precision_score(y_train, yhat_train)
  prec_test = precision_score(y_test, yhat_test)

  return prec_train, prec_test
```

```python
k = [2, 3, 4, 5]
prec_train = []
prec_test = []

for i in k:
  train, test = evaluate_k(i, x_train, y_train, x_test, y_test)
  prec_train.append(train)
  prec_test.append(test)
```

```python
plt.plot(k, prec_train, label='Treinamento')
plt.plot(k, prec_test, label='Teste')
plt.legend()
plt.show()
```

## Regressão Linear para Predição Contínua (Emissão de CO₂)

A **Regressão Linear** é um dos algoritmos mais fundamentais e amplamente utilizados em Machine Learning para prever resultados contínuos. Diferente da classificação (onde prevemos categorias discretas como 'risco' ou 'sem risco'), a regressão lida com a previsão de valores numéricos que podem cair em um intervalo contínuo, como preço de casas, temperatura ou, no nosso caso, **emissão de CO₂**.

### Como Funciona?

Imagine que você quer prever a quantidade de CO₂ que um veículo emitirá com base em suas características, como o tamanho do motor, o consumo de combustível e o peso. A regressão linear tenta encontrar a "melhor linha reta" que descreve a relação entre essas características (variáveis independentes ou *features*) e a emissão de CO₂ (a variável dependente ou *target*).

Matematicamente, um modelo de regressão linear simples (com uma única *feature*) pode ser representado como:

$$ y = \beta_0 + \beta_1x + \epsilon $$

Onde:
- $y$ é o valor que queremos prever (ex: emissão de CO₂).
- $x$ é a variável de entrada (ex: tamanho do motor).
- $\beta_0$ é o intercepto (o valor de $y$ quando $x$ é zero).
- $\beta_1$ é o coeficiente da variável $x$, indicando o quanto $y$ muda para cada unidade de mudança em $x$.
- $\epsilon$ representa o erro irredutível ou o ruído.

Para múltiplas *features*, a equação se expande para:

$$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n + \epsilon $$

O objetivo do algoritmo de regressão linear é encontrar os valores ótimos para $\beta_0, \beta_1, ..., \beta_n$ que minimizem a diferença entre os valores previstos pelo modelo e os valores reais. Essa "melhor linha de ajuste" (ou plano/hiperplano em dimensões superiores) é geralmente encontrada minimizando a soma dos quadrados dos erros (distâncias verticais entre os pontos de dados e a linha), um método conhecido como **Mínimos Quadrados Ordinários (MQO)**.

### Analogia com Emissão de CO₂

No nosso contexto, a Regressão Linear nos ajudará a entender como diferentes atributos de um veículo contribuem para sua emissão de CO₂. Por exemplo, se descobrirmos que um aumento de 1 litro no tamanho do motor corresponde a um aumento de 20 g/km na emissão de CO₂, isso é uma informação valiosa para fabricantes e reguladores.

```python
# caminho para o arquivo
PATH='/content/drive/MyDrive/Pos_Graduacao/FundamentosMachineLearning/data/co2.csv'
data = pd.read_csv(PATH)
data.head()
```

```python
# verificando dados nulos
display(data.isna().sum())
```

```python
data.info()
```

```python
data['Fuel Consumption Comb (L/100 km)'].dtype
```

```python
# colinearidade
numeric_features = data.select_dtypes(include=[np.number]).columns
display(data[numeric_features].corr(method='spearman'))
```

```python
# selecionando features
target = 'CO2 Emissions(g/km)'
categ_feat = ['Vehicle Class']
num_feat = [c for c in numeric_features if c != target]
```

```python
plt.plot(data['Fuel Consumption Comb (L/100 km)'], data[target], 'k.')
```

```python
# compor nosso pipeline reg
col_transformer2 = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), categ_feat)
], remainder='passthrough')

reg_pipeline = Pipeline([
    ('col_transform', col_transformer2),
    ('selector', SelectKBest(k=3, score_func=f_regression)),
    ('scaler', RobustScaler()),
    ('model', LinearRegression())
])
```

```python
# separação de treino e teste
x = data[num_feat + categ_feat]
y = data[[target]]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=RANDOM_STATE)
```

```python
# treinamento do pipeline
reg_pipeline.fit(x_train, y_train)
```

```python
# calculando previsões
yhat_train = reg_pipeline.predict(x_train)
yhat_test = reg_pipeline.predict(x_test)

print('R2 Treinamento:', r2_score(y_train, yhat_train))
print('R2 Teste:', r2_score(y_test, yhat_test))
```

```python
# acessar os coeficientes
print(reg_pipeline['model'].intercept_)
print(reg_pipeline['model'].coef_)
```

```python
plt.plot(y_test, y_test, 'r-', label='Ideal')
plt.plot(y_test, yhat_test, '.k', label='Predicted')
plt.xlabel('Real Values')
plt.ylabel('Predicted Values')
plt.legend()
```

### Comparação e Resumo dos Modelos Lineares: Perceptron, Regressão Logística e Regressão Linear

Todos os três modelos explorados – **Perceptron**, **Regressão Logística** e **Regressão Linear** – compartilham um princípio fundamental no Machine Learning: eles buscam encontrar uma relação linear nos dados para fazer previsões ou classificações. No entanto, a forma como aplicam essa linearidade e o tipo de problema que resolvem são distintos.

#### Princípio Comum: Funções de Decisão Lineares

Em sua essência, todos esses modelos utilizam uma combinação linear das características de entrada (features) para formar sua decisão ou predição. Isso significa que eles calculam uma soma ponderada dos inputs, onde cada input é multiplicado por um peso e um viés (intercepto) é adicionado:

$$ z = w_0 + w_1x_1 + w_2x_2 + ... + w_nx_n $$

O que difere é o que acontece com este valor $z$ em cada modelo:

*   **Perceptron:** Usa $z$ diretamente para uma decisão binária rígida via uma função degrau.
*   **Regressão Logística:** Passa $z$ por uma função sigmoide para obter uma probabilidade.
*   **Regressão Linear:** Usa $z$ como a própria predição contínua (com $z$ sendo o $y$ predito).

#### Diferenças em Aplicações e Saídas

| Característica              | Perceptron                                    | Regressão Logística                                | Regressão Linear                                    |
| :-------------------------- | :-------------------------------------------- | :------------------------------------------------- | :-------------------------------------------------- |
| **Tipo de Problema**        | Classificação Binária (Linearmente Separável) | Classificação Binária (ou Multiclasse)             | Regressão (Predição Contínua)                       |
| **Saída**                   | Classe discreta (0 ou 1)                      | Probabilidade de pertencer à classe positiva (0 a 1) | Valor numérico contínuo                             |
| **Função de Ativação/Link** | Função degrau (Heaviside)                     | Função Sigmoide                                    | Nenhuma (Função Identidade)                         |
| **Analogia**                | Portão "sim/não" simples                    | Estimador de "chance"                           | Encontrar a "linha de melhor ajuste"              |

#### Quando Usar Cada Modelo e Sua Interpretibilidade

*   **Perceptron:**
    *   **Quando usar:** Raramente usado isoladamente hoje em dia devido à sua limitação a dados linearmente separáveis e à falta de saída probabilística. Serve como um bloco construtor teórico para redes neurais. Pode ser útil como um classificador de *baseline* muito simples em cenários específicos onde a interpretabilidade de uma decisão binária clara é prioritária e a separabilidade linear é garantida.
    *   **Interpretibilidade:** Alta, mas limitada. Os pesos indicam a importância e a direção de cada feature na decisão, mas a saída é binária e não oferece nuances de confiança.

*   **Regressão Logística:**
    *   **Quando usar:** Ideal para problemas de classificação binária (e pode ser estendida para multiclasse) onde a saída probabilística é desejada. Por exemplo, prever a probabilidade de um cliente abandonar um serviço, a probabilidade de um e-mail ser spam, ou o risco de inadimplência. É robusta, eficiente e uma ótima escolha para *baselines* de classificação.
    *   **Interpretibilidade:** Alta. Os coeficientes podem ser interpretados como a mudança no *log-odds* da classe positiva para cada unidade de aumento na feature, mantendo outras constantes. Isso permite entender a direção e a força da relação entre as features e a probabilidade do evento.

*   **Regressão Linear:**
    *   **Quando usar:** Perfeita para prever um valor numérico contínuo, como o preço de uma casa, a temperatura do próximo mês, o consumo de combustível de um carro ou as emissões de CO₂. É o ponto de partida para qualquer problema de regressão devido à sua simplicidade e interpretabilidade.
    *   **Interpretibilidade:** Muito alta. Os coeficientes indicam diretamente o quanto a variável dependente (target) muda para cada unidade de aumento na feature correspondente, mantendo outras features constantes. Isso fornece insights claros sobre a influência de cada fator na predição.
