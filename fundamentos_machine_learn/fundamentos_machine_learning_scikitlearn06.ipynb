
# Introdução às Métricas de Machine Learning

Métricas de avaliação em Machine Learning são ferramentas quantitativas usadas para medir o desempenho de um modelo preditivo. Elas fornecem uma maneira objetiva de entender quão bem um modelo está realizando sua tarefa, seja ela classificação, regressão ou agrupamento, ao comparar as previsões do modelo com os valores reais ou esperados.

A importância dessas métricas é multifacetada e crucial para o desenvolvimento e a implantação de modelos de ML eficazes. Primeiramente, elas permitem **quantificar o desempenho de um modelo**, transformando a avaliação de 'bom' ou 'ruim' em números concretos. Em segundo lugar, são indispensáveis para **comparar diferentes modelos ou diferentes configurações do mesmo modelo**, ajudando a identificar qual abordagem é superior para um determinado problema. Essa comparação é vital para a otimização de modelos. Além disso, as métricas são fundamentais para **tomar decisões informadas sobre a implantação de modelos em cenários reais**, pois garantem que apenas modelos com desempenho aceitável sejam colocados em produção. Finalmente, elas ajudam a **entender os pontos fortes e fracos de um modelo em relação ao problema de negócio**, permitindo que os desenvolvedores ajustem e melhorem o modelo para atender às necessidades específicas do contexto em que será utilizado.

## Métricas de Classificação: Matriz de Confusão

A **Matriz de Confusão** é uma ferramenta fundamental para avaliar o desempenho de um algoritmo de classificação, especialmente em problemas de classificação binária. Ela é uma tabela que resume os resultados previstos por um modelo de classificação em comparação com os resultados reais. Através dela, podemos visualizar rapidamente onde o modelo acertou e onde errou, fornecendo uma base para o cálculo de diversas métricas de avaliação.

### Componentes da Matriz de Confusão:

A matriz é composta por quatro elementos principais:

*   **Verdadeiros Positivos (True Positives - TP)**: Representam os casos em que o modelo previu corretamente a classe positiva. Ou seja, o modelo classificou como positivo, e a classe real era, de fato, positiva.

*   **Verdadeiros Negativos (True Negatives - TN)**: Representam os casos em que o modelo previu corretamente a classe negativa. O modelo classificou como negativo, e a classe real era, de fato, negativa.

*   **Falsos Positivos (False Positives - FP)**: Conhecidos como Erro Tipo I, são os casos em que o modelo previu a classe positiva, mas a classe real era negativa. O modelo "alarmou" falsamente.

*   **Falsos Negativos (False Negatives - FN)**: Conhecidos como Erro Tipo II, são os casos em que o modelo previu a classe negativa, mas a classe real era positiva. O modelo "perdeu" um caso positivo que deveria ter sido identificado.

## Precisão e Recall

### Definição de Precisão (Precision)

**Precisão** mede a proporção de verdadeiros positivos (TP) entre todos os exemplos que foram classificados como positivos (TP + Falsos Positivos - FP). Em outras palavras, ela nos diz, dos itens que o modelo identificou como relevantes, quantos realmente eram relevantes. É crucial em cenários onde o custo de um falso positivo é alto.

**Fórmula de Precisão:**

$$ \text{Precisão} = \frac{TP}{TP + FP} $$


### Definição de Recall (Sensibilidade ou Cobertura)

**Recall** mede a proporção de verdadeiros positivos (TP) entre todos os exemplos que são realmente positivos (TP + Falsos Negativos - FN). Ou seja, ela nos diz, de todos os itens realmente relevantes, quantos o modelo conseguiu identificar. É crucial em cenários onde o custo de um falso negativo é alto.

**Fórmula de Recall:**

$$ \text{Recall} = \frac{TP}{TP + FN} $$


### Trade-off entre Precisão e Recall

Existe um **trade-off inerente** entre Precisão e Recall na maioria dos modelos de classificação. Geralmente, ao tentar otimizar um, o outro tende a diminuir.

*   **Aumentar a Precisão** significa que o modelo se torna mais conservador, classificando um exemplo como positivo apenas quando tem alta certeza. Isso reduz o número de falsos positivos, mas pode aumentar o número de falsos negativos, diminuindo o Recall.
*   **Aumentar o Recall** significa que o modelo se torna mais abrangente, tentando capturar o máximo de exemplos positivos possível, mesmo que isso signifique incluir alguns falsos positivos. Isso reduz o número de falsos negativos, mas pode aumentar o número de falsos positivos, diminuindo a Precisão.

**Exemplo Prático: Diagnóstico Médico de Doença Rara**

Imagine um modelo de Machine Learning desenvolvido para diagnosticar uma doença rara e grave.

*   **Priorizar Precisão:** Se o hospital prioriza a **Precisão**, o modelo será ajustado para ter poucos falsos positivos. Isso significa que, se o modelo diagnosticar alguém com a doença, é quase certo que essa pessoa realmente a tem. No entanto, o modelo pode ignorar alguns casos reais (falsos negativos), resultando em um baixo Recall. Neste cenário, o custo de um falso positivo (tratar uma pessoa saudável com medicação agressiva) é muito alto.

*   **Priorizar Recall:** Se o hospital prioriza o **Recall**, o modelo será ajustado para identificar o máximo de pacientes com a doença possível, mesmo que isso signifique diagnosticar erroneamente algumas pessoas saudáveis (falsos positivos). Neste caso, o Recall será alto, mas a Precisão será menor. Aqui, o custo de um falso negativo (não diagnosticar alguém com a doença grave) é considerado muito mais alto, pois o atraso no tratamento pode ser fatal.

Neste exemplo, a escolha entre priorizar Precisão ou Recall depende diretamente das consequências dos erros do modelo, refletindo o trade-off fundamental entre essas duas métricas.

### F1-score

O F1-score é uma métrica que representa a média harmônica entre a precisão (Precision) e o recall (Recall). Ele é particularmente útil quando se busca um equilíbrio entre falsos positivos e falsos negativos, ou seja, quando ambos os tipos de erros têm custos semelhantes. Uma alta pontuação F1-score indica que o modelo tem tanto alta precisão quanto alto recall.

Sua fórmula em LaTeX é:

$$\text{F1-score} = 2 \cdot \frac{\text{Precisão} \cdot \text{Recall}}{\text{Precisão} + \text{Recall}}$$

ou, em termos de Verdadeiros Positivos (TP), Falsos Positivos (FP) e Falsos Negativos (FN):

$$\text{F1-score} = \frac{2TP}{2TP + FP + FN}$$

### F-beta score

O F-beta score é uma generalização do F1-score que permite ponderar a importância da precisão ou do recall, por meio do parâmetro $\beta$ (beta). Ele é útil quando os custos de falsos positivos e falsos negativos não são iguais e você deseja dar mais peso a uma das métricas.

#### O parâmetro $\beta$

O valor de $\beta$ determina a influência relativa do recall em relação à precisão:

*   **Quando $\beta = 1$**: O F-beta score se torna o F1-score, dando peso igual à precisão e ao recall. Isso significa que falsos positivos e falsos negativos são considerados igualmente importantes.

*   **Quando $\beta > 1$**: Dá mais peso ao recall. Por exemplo, um F2-score ($\beta = 2$) significa que o recall é considerado duas vezes mais importante que a precisão. Isso é útil em situações onde é crucial minimizar os falsos negativos (ou seja, não perder instâncias positivas), mesmo que isso signifique um aumento nos falsos positivos. Exemplos incluem a detecção de doenças graves (não se quer deixar de detectar um paciente doente).

*   **Quando $\beta < 1$**: Dá mais peso à precisão. Por exemplo, um F0.5-score ($\beta = 0.5$) significa que a precisão é considerada duas vezes mais importante que o recall. Isso é útil em situações onde é crítico minimizar os falsos positivos, mesmo que isso signifique perder algumas instâncias positivas. Exemplos incluem a detecção de spam (não se quer classificar um e-mail importante como spam) ou sistemas de recomendação (não se quer recomendar algo que o usuário realmente não goste).

Sua fórmula em LaTeX é:

$$\text{F}_\beta\text{-score} = (1 + \beta^2) \cdot \frac{\text{Precisão} \cdot \text{Recall}}{(\beta^2 \cdot \text{Precisão}) + \text{Recall}}$$

### Cenários Práticos

A escolha do valor de $\beta$ depende do problema de negócio e do custo associado a cada tipo de erro:

*   **Detecção de doenças (graves)**: Preferimos um $\beta > 1$ (e.g., F2-score). É mais importante identificar todos os casos positivos (alto recall) para que recebam tratamento, mesmo que alguns resultados sejam falsos positivos (pessoas saudáveis que são encaminhadas para exames adicionais). O custo de um falso negativo (não detectar a doença) é muito alto.

*   **Detecção de spam**: Preferimos um $\beta < 1$ (e.g., F0.5-score). É crucial ter alta precisão para evitar que e-mails legítimos sejam marcados como spam (falsos positivos). Perder um ou outro spam (falsos negativos) é menos problemoso do que perder um e-mail importante.

*   **Recomendação de produtos**: Similar à detecção de spam, um $\beta < 1$ pode ser preferível. Queremos ter certeza de que os produtos recomendados são realmente relevantes (alta precisão), para não frustrar o usuário. Perder algumas boas recomendações (falsos negativos) pode ser aceitável.

Entender o trade-off entre precisão e recall e como o parâmetro $\beta$ ajusta esse equilíbrio é fundamental para a avaliação eficaz de modelos de classificação em diversos contextos.

## ROC AUC (Receiver Operating Characteristic - Area Under Curve)

### O que é a Curva ROC?
A Curva ROC (Receiver Operating Characteristic) é uma ferramenta gráfica essencial para avaliar o desempenho de modelos de classificação binária em diferentes **limiares (thresholds)** de probabilidade. Ela ilustra a capacidade discriminatória do modelo, ou seja, quão bem ele é capaz de distinguir entre as classes positiva e negativa.

image.png

### Eixos da Curva ROC
A curva ROC é plotada com dois eixos principais:

*   **True Positive Rate (TPR)**: Também conhecido como **Recall** ou **Sensibilidade**, o TPR mede a proporção de **positivos reais** que foram corretamente identificados pelo modelo. Em outras palavras, é a capacidade do modelo de encontrar todas as instâncias positivas.

    A fórmula do TPR é:
    $$ \text{TPR (Recall)} = \frac{TP}{TP + FN} $$

    Onde:
    *   `TP` (True Positives): Casos positivos corretamente classificados como positivos.
    *   `FN` (False Negatives): Casos positivos incorretamente classificados como negativos.

*   **False Positive Rate (FPR)**: O FPR mede a proporção de **negativos reais** que foram incorretamente classificados como positivos. É uma medida de quantos resultados falsos positivos o modelo gera.

    A fórmula do FPR é:
    $$ \text{FPR} = \frac{FP}{FP + TN} $$

    Onde:
    *   `FP` (False Positives): Casos negativos incorretamente classificados como positivos.
    *   `TN` (True Negatives): Casos negativos corretamente classificados como negativos.

### Construção da Curva ROC
A curva ROC é construída variando-se o **limiar de decisão (threshold)** do classificador. Um classificador binário normalmente produz uma probabilidade para a classe positiva. O limiar determina o ponto de corte: se a probabilidade prevista for maior que o limiar, o caso é classificado como positivo; caso contrário, é classificado como negativo.

Cada ponto na curva ROC corresponde a um par (FPR, TPR) obtido para um determinado limiar de probabilidade. Ao mover o limiar de 0 a 1, geramos uma série de pares (FPR, TPR) que, quando plotados, formam a curva ROC.

### Interpretação da Curva ROC
*   **Curva próxima ao canto superior esquerdo**: Um modelo ideal teria um TPR de 1 (todos os positivos corretamente identificados) e um FPR de 0 (nenhum falso positivo). Portanto, uma curva ROC que se aproxima do canto superior esquerdo (0,1) do gráfico indica um **melhor desempenho** do modelo, significando que ele consegue distinguir bem entre as classes.

*   **Linha diagonal (random classifier)**: A linha diagonal que vai do canto inferior esquerdo (0,0) ao canto superior direito (1,1) representa um classificador aleatório ou um modelo que não tem poder discriminatório. Para este modelo, a chance de classificar corretamente um positivo é a mesma que a de classificar incorretamente um negativo.

### Área Sob a Curva (AUC - Area Under the Curve)
A AUC é uma métrica agregada que quantifica o desempenho geral de um modelo de classificação. Ela representa a **área total sob a curva ROC**. Um valor de AUC pode ser interpretado como a probabilidade de que o modelo classifique um par aleatório de exemplos (um positivo e um negativo) corretamente, ou seja, que ele atribua uma probabilidade maior ao exemplo positivo do que ao exemplo negativo.

### Interpretação do Valor do AUC
*   **AUC = 1.0**: Indica um modelo perfeito, capaz de distinguir perfeitamente entre as classes positiva e negativa. O modelo tem 100% de TPR e 0% de FPR para algum limiar.
*   **AUC = 0.5**: Indica um modelo que se comporta de forma aleatória, não sendo melhor do que um chute aleatório na distinção das classes. Isso corresponde à linha diagonal na curva ROC.
*   **AUC < 0.5**: Indica um modelo que é pior do que um chute aleatório. Em tais casos, pode-se simplesmente inverter as previsões do modelo para obter um desempenho melhor que 0.5.

### Importância do AUC
A AUC é uma métrica muito valorizada porque é **independente do limiar de classificação** escolhido. Isso a torna uma excelente métrica única para comparar diferentes modelos de classificação, pois ela avalia a capacidade discriminatória intrínseca do modelo em todos os possíveis limiares, ao invés de depender de um ponto de corte específico que pode ser arbitrário ou otimizado para uma situação particular.

## Curva Precision-Recall

A Curva Precision-Recall (P-R Curve) é uma ferramenta visual que plota a **Precisão** (Precision) versus o **Recall** para diferentes limiares (thresholds) de probabilidade de um modelo de classificação. Ela é particularmente útil para avaliar o desempenho de modelos em problemas de classificação binária, especialmente quando se lida com datasets **desbalanceados**.

image.png

### Uso em Datasets Desbalanceados
Em datasets onde uma classe é significativamente menos representada que a outra (datasets desbalanceados), métricas como acurácia ou a Curva Receiver Operating Characteristic (ROC) podem ser enganosas. Um modelo pode atingir alta acurácia simplesmente classificando a maioria das instâncias como a classe majoritária, ignorando a classe minoritária, que muitas vezes é a classe de interesse (e.g., detecção de fraudes, doenças raras).

A Curva Precision-Recall foca no desempenho da classe positiva. A Precisão mede a proporção de verdadeiros positivos entre todas as previsões positivas (TP / (TP + FP)), enquanto o Recall (ou Sensibilidade) mede a proporção de verdadeiros positivos entre todas as instâncias positivas reais (TP / (TP + FN)). Ao focar nesses dois aspectos, a P-R Curve oferece uma visão mais realista do desempenho do modelo em identificar a classe minoritária, revelando se o modelo é capaz de encontrar a maioria das instâncias positivas sem muitas falsas detecções.

### Impacto da Mudança do Threshold de Probabilidade
A maioria dos modelos de classificação gera uma probabilidade de uma instância pertencer à classe positiva. Um limiar (threshold) é então aplicado a essa probabilidade para decidir a classificação final (e.g., se a probabilidade > 0.5, classifica como positiva). A alteração desse limiar impacta diretamente a relação entre Precisão e Recall:

*   **Aumentando o Threshold:** Se aumentarmos o limiar de probabilidade (e.g., de 0.5 para 0.8), o modelo se tornará mais "cauteloso" para classificar algo como positivo. Isso geralmente leva a:
    *   **Maior Precisão:** Menos falsos positivos, pois o modelo só classifica como positivo quando há alta certeza. No entanto, pode ignorar alguns verdadeiros positivos, resultando em...
    *   **Menor Recall:** Mais falsos negativos, pois o modelo perde instâncias positivas que tinham probabilidades abaixo do novo e mais alto limiar.

*   **Diminuindo o Threshold:** Se diminuirmos o limiar de probabilidade (e.g., de 0.5 para 0.2), o modelo se tornará mais "agressivo" para classificar algo como positivo. Isso geralmente leva a:
    *   **Maior Recall:** Mais verdadeiros positivos serão capturados, pois o modelo aceita probabilidades mais baixas. No entanto, isso pode vir com um custo de...
    *   **Menor Precisão:** Mais falsos positivos, pois o modelo classifica erroneamente instâncias negativas como positivas.

Cada ponto na Curva Precision-Recall representa um par (Recall, Precisão) correspondente a um limiar de probabilidade diferente. Ao mover o limiar, o trade-off entre Precisão e Recall se torna evidente. Uma curva que se mantém próxima ao canto superior direito (onde Precisão e Recall são ambos altos, ou (1,1)) indica um modelo de melhor desempenho. A **Área Sob a Curva Precision-Recall (AUPRC)** é uma métrica de resumo que quantifica o desempenho geral do modelo através de todos os possíveis limiares, sendo um valor mais alto indicativo de um melhor modelo.

## Métricas de Regressão: Conceitos Iniciais

### Necessidade de Métricas Específicas para Regressão

Em Machine Learning, é crucial diferenciar entre problemas de **classificação** e **regressão**. Enquanto problemas de classificação visam prever uma categoria (e.g., sim/não, gato/cachorro), problemas de regressão buscam prever um valor contínuo (e.g., preço de uma casa, temperatura).

As métricas utilizadas para avaliar modelos de classificação (como Acurácia, Precisão, Recall, F1-Score) não são adequadas para modelos de regressão. Isso ocorre porque as métricas de classificação medem a concordância exata ou a capacidade de diferenciar entre classes discretas. Em regressão, raramente se espera uma previsão exata, mas sim uma previsão que esteja "próxima" do valor real.

As métricas de regressão, por outro lado, focam em quantificar a **'distância'** ou **'diferença'** entre os valores previstos ($\hat{y}$) e os valores reais (y), ou seja, o **erro** do modelo. Elas avaliam a qualidade do ajuste do modelo aos dados, buscando minimizá-las para obter um melhor desempenho preditivo. Quanto menor o valor da métrica de erro (na maioria dos casos), melhor o modelo se ajusta aos dados e mais precisas são suas previsões.

### MAE (Mean Absolute Error)

O MAE (Mean Absolute Error) é uma métrica utilizada para medir a precisão de um modelo de previsão. Ele representa a média das diferenças absolutas entre os valores previstos pelo modelo e os valores reais.

### Fórmula do MAE

A fórmula matemática para o MAE é:

$$ MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$

Onde:
- $y_i$ são os valores reais.
- $\hat{y}_i$ são os valores previstos pelo modelo.
- $n$ é o número total de observações.

### Interpretabilidade do MAE

O MAE é uma métrica de fácil interpretabilidade porque representa a magnitude média do erro em unidades da variável dependente. Por exemplo, se o MAE é de 5 unidades, significa que, em média, as previsões do modelo estão erradas em 5 unidades. Ele fornece uma compreensão direta de quão grandes são os erros médios sem que o sinal dos erros (superestimação ou subestimação) afete o resultado.

### Sensibilidade a Outliers

O MAE é menos sensível a outliers do que outras métricas como o RMSE (Root Mean Squared Error) porque trata os erros de forma linear. Diferenças maiores entre valores previstos e reais contribuem proporcionalmente para o MAE. Isso significa que erros grandes (outliers) não são penalizados de forma quadrática (como no RMSE, onde o erro é elevado ao quadrado), tornando o MAE uma escolha mais robusta quando a presença de outliers é uma preocupação e você não quer que eles dominem a métrica de erro.

### RMSE (Root Mean Squared Error)

O RMSE (Root Mean Squared Error) é uma métrica comumente usada para avaliar a precisão de um modelo de previsão. Ele mede a raiz quadrada da média dos erros ao quadrado entre os valores previstos ($\hat{y}_i$) e os valores reais ($y_i$).

A fórmula do RMSE é apresentada abaixo:

$$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

Onde:
- $y_i$ são os valores reais (observados)
- $\hat{y}_i$ são os valores previstos (pelo modelo)
- $n$ é o número total de observações ou pontos de dados

#### Comparação com MAE (Mean Absolute Error)

O RMSE e o MAE (Mean Absolute Error) são ambas métricas para medir o erro de previsão, mas diferem na forma como penalizam os erros. O RMSE penaliza erros maiores de forma mais significativa do que o MAE, devido ao termo ao quadrado em sua fórmula. Isso significa que erros grandes contribuem desproporcionalmente mais para o RMSE do que para o MAE.

Por exemplo, um erro de 10 unidades contribui com 100 para a soma dos quadrados no RMSE, enquanto no MAE contribui apenas com 10. Esta característica torna o RMSE mais sensível a **outliers** (valores atípicos ou anômalos) nos dados. Modelos que minimizam o RMSE tentarão reduzir esses erros grandes, mesmo que isso signifique fazer pequenos erros em muitos outros pontos de dados.

Devido ao termo ao quadrado e à subsequente raiz quadrada, o valor do RMSE é geralmente maior ou igual ao MAE, a menos que todos os erros sejam zero ou um.

### R2 (Coefficient of Determination)

O $R^2$, ou Coeficiente de Determinação, é uma medida estatística que representa a proporção da variância na variável dependente que pode ser explicada pelas variáveis independentes em um modelo de regressão. Em termos mais simples, ele indica o quão bem o modelo se ajusta aos dados observados.

#### Fórmula do R2:

A fórmula do R2 é dada por:

$$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$

Onde:
*   $SS_{res}$ (Soma dos Quadrados dos Resíduos) é a soma dos quadrados das diferenças entre os valores observados e os valores previstos pelo modelo.
*   $SS_{tot}$ (Soma Total dos Quadrados) é a soma dos quadrados das diferenças entre os valores observados e a média da variável dependente.

#### Como o R2 mede a proporção da variância:

O $R^2$ varia de 0 a 1 (ou 0% a 100% quando expresso em porcentagem). Ele explica como a variância da variável dependente é previsível a partir das variáveis independentes:

*   **$R^2 = 0$**: Isso indica que o modelo de regressão não explica nenhuma da variabilidade da variável dependente em torno de sua média. Em outras palavras, as variáveis independentes não fornecem nenhuma informação útil para prever a variável dependente.

*   **$R^2 = 1$**: Isso significa que o modelo de regressão explica toda a variabilidade da variável dependente. Todos os pontos de dados caem perfeitamente na linha de regressão, indicando um ajuste perfeito.

*   **$0 < R^2 < 1$**: Este é o cenário mais comum. Um $R^2$ de, por exemplo, 0.75 (ou 75%) indica que 75% da variância na variável dependente pode ser explicada pelas variáveis independentes incluídas no modelo, e os restantes 25% são devidos a fatores não explicados (resíduos). Quanto mais próximo de 1 o valor de $R^2$ estiver, melhor será o ajuste do modelo aos dados e maior será a capacidade do modelo de prever a variável dependente.

### MAPE (Mean Absolute Percentage Error)

O MAPE (Mean Absolute Percentage Error) é uma métrica de avaliação de precisão de previsão que mede o tamanho médio do erro de previsão em termos percentuais. Ele expressa a precisão como uma porcentagem do erro em relação aos valores reais, o que o torna útil para entender o erro em termos relativos, independentemente da escala dos dados.

### Fórmula do MAPE

A fórmula para calcular o MAPE é a seguinte:

$$\text{MAPE} = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100\%$$

Onde:
- $y_i$ são os valores reais (observados).
- $\hat{y}_i$ são os valores previstos pelo modelo.
- $n$ é o número total de observações.

### Utilidade do MAPE

A principal utilidade do MAPE reside na sua **interpretabilidade intuitiva**. Por ser uma medida percentual, o MAPE é fácil de entender e comunicar, mesmo para públicos não técnicos. Ele permite comparar a precisão de previsões em diferentes séries temporais ou conjuntos de dados que possuem escalas distintas, pois o erro é normalizado pelo valor real. Um MAPE de 10%, por exemplo, indica que o erro médio é de 10% do valor real, o que é uma informação valiosa para a tomada de decisões.

### Limitações do MAPE

Apesar de sua utilidade, o MAPE possui algumas limitações importantes:

1.  **Divisão por zero:** O MAPE não pode ser calculado ou gera um valor infinito quando os valores reais ($y_i$) são iguais a zero. Isso ocorre porque a fórmula envolve uma divisão por $y_i$. Este é um problema comum em dados com valores esparsos ou onde a demanda pode ocasionalmente ser zero.
2.  **Instabilidade com valores próximos de zero:** Quando os valores reais ($y_i$) são muito próximos de zero, mesmo um pequeno erro absoluto pode resultar em um erro percentual muito grande (e.g., um erro de 1 para um valor real de 0.01 resulta em 10000% de erro). Isso pode distorcer a média e tornar o MAPE enganoso.
3.  **Assimetria em relação aos erros:** O MAPE dá mais peso aos erros negativos (quando a previsão é maior que o valor real, ou seja, superestimações) do que aos erros positivos (quando a previsão é menor que o valor real, ou seja, subestimações). Isso significa que superestimar a demanda tem um impacto maior no MAPE do que subestimá-la em uma proporção similar, o que pode enviesar a otimização de modelos que usam o MAPE como função de custo.
